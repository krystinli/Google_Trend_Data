### Interest over time - "House Price Toronto vs. Vancouver Google Search"
1. read data + assign column names 
2. explore the data
3. plot the data

Numbers represent search interest relative to the highest point on the chart for the given region and time. A value of 100 is the peak popularity for the term. A value of 50 means that the term is half as popular. A score of 0 means there was not enough data for this term.

```python
import numpy as np
import pandas as pd
import datetime

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.dates as mdates
%matplotlib inline 
```

```python
# https://trends.google.com/trends/explore?date=all&geo=CA&q=house%20price%20Toronto,house%20price%20Vancouver

data = pd.read_csv("house_price.csv")
data.head()
```

```python
# step 1): reload data by setting the right header 
data = pd.read_csv("house_price.csv", header=1)
data.rename(columns={"house price Toronto: (Canada)":"Toronto", "house price Vancouver: (Canada)":"Vancouver"}, inplace=True)
data.head()
```

```python
# step 2): explore the data
data.info()
```

```python
data.describe()
```

```python
# step 3): simple plot 
plt.plot(data.Month, data.Toronto, data.Vancouver)
plt.show()
```

```python
# step 4): better plot 
fig, ax = plt.subplots()
ax.plot("Month", "Toronto", data=data)
ax.plot("Month", "Vancouver", data=data)

# labels and legend 
ax.set(xlabel='Month', ylabel='Interest Score', title='Google Search of Toronto vs. Vancouver House Price')
plt.legend()

# rotate x-axis and set max num of ticks displayed 
plt.xticks(data.index, rotation=50)
ax.xaxis.set_major_locator(plt.MaxNLocator(45)) 

# resize img 
fig.set_size_inches(18, 5)
plt.show()
```
